```python
import cv2
import os
import numpy as np
input_image = '../../2D_Landmarks/Dataset/H3DS_sample/sample2/256/'
input_features = []
for file in os.listdir(input_image):
    curr_input = os.path.join(input_image, file)
    curr_img = cv2.imread(curr_input)
    curr_img = curr_img.flatten()
    input_features.append(curr_img)
```

```python
input_features = np.array(input_features)
print(input_features.shape)
```

```python
from sklearn.decomposition import PCA

################# Your answer for Question 3.1   #################

# name the transformed (dimension reduced) feature as projected
pca = PCA(2)
pca.fit(input_features)
projected = pca.transform(input_features)
```
